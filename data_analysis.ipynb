```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset_cleaned = pd.read_csv('dataset_cleaned.csv')
dataset_cleaned.head()
```

```python
# Ограничение данных для построения графика
dataset_cleaned_filtered = dataset_cleaned[dataset_cleaned['Quantity'] < 100]

dataset_cleaned_filtered['Quantity'].hist(bins=20)
plt.title('Распределение количества товаров в заказе (до 100)')
plt.xlabel('Количество товаров')
plt.ylabel('Частота')
plt.show()
```

```python
dataset_cleaned['UnitPrice'].hist(bins=20)
plt.title('Распределение цен товаров')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

```python
dataset_cleaned['InvoiceDate'] = pd.to_datetime(dataset_cleaned['InvoiceDate'])
monthly_sales = dataset_cleaned.groupby(dataset_cleaned['InvoiceDate'].dt.to_period('M')).size()

monthly_sales.plot(kind='line')
plt.title('Количество заказов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество заказов')
plt.show()
```
