```python
import pandas as pd
import matplotlib.pyplot as plt


dataset_cleaned = pd.read_csv('dataset_cleaned.csv')

dataset_cleaned['InvoiceDate'] = pd.to_datetime(dataset_cleaned['InvoiceDate'])

dataset_cleaned.head()
```

```python
# по месяцам
monthly_sales = dataset_cleaned.groupby(dataset_cleaned['InvoiceDate'].dt.to_period('M')).size()

monthly_sales.plot(kind='line', marker='o')
plt.title('Количество заказов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество заказов')
plt.savefig('images/monthly_sales.png')
plt.show()
```

```python
# Добавление столбца с днем недели
dataset_cleaned['day_of_week'] = dataset_cleaned['InvoiceDate'].dt.day_name()

# Группировка данных по дням недели
weekday_sales = dataset_cleaned.groupby('day_of_week').size().sort_values()

# Построение графика
weekday_sales.plot(kind='bar')
plt.title('Количество заказов по дням недели')
plt.xlabel('День недели')
plt.ylabel('Количество заказов')
plt.show()
```

```python
# Добавление столбца с часом заказа
dataset_cleaned['hour'] = dataset_cleaned['InvoiceDate'].dt.hour

# Группировка данных по часам
hourly_sales = dataset_cleaned.groupby('hour').size()

# Построение графика
hourly_sales.plot(kind='line', marker='o')
plt.title('Количество заказов по времени суток')
plt.xlabel('Час')
plt.ylabel('Количество заказов')
plt.show()
```
