```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset_cleaned = pd.read_csv('dataset_cleaned.csv')
dataset_cleaned.head()
```

```python

dataset_cleaned_filtered = dataset_cleaned[dataset_cleaned['Quantity'] < 100]

dataset_cleaned_filtered['Quantity'].hist(bins=20)
plt.title('Distribution of the number of items in an order (up to 100)')
plt.xlabel('Quantity of goods')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset_cleaned['UnitPrice'].hist(bins=20)
plt.title('Distribution of commodity prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset_cleaned['InvoiceDate'] = pd.to_datetime(dataset_cleaned['InvoiceDate'])
monthly_sales = dataset_cleaned.groupby(dataset_cleaned['InvoiceDate'].dt.to_period('M')).size()

monthly_sales.plot(kind='line')
plt.title('Number of orders by month')
plt.xlabel('Month')
plt.ylabel('Number of orders')
plt.show()
```
